```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("GM19_tpm.txt", sep="\t", index_col="Geneid")
```

```python
df
```

```python
dflog = np.log2(df)
dflog
```

```python
dfFold = dflog.pct_change(axis='columns')
dfFold.dropna(axis=1, inplace=True)
dfFold
```

```python
dfDGEup = dfFold[dfFold >= 0.5].dropna(thresh=1)
dfDGEdown = dfFold[dfFold <= -0.5].dropna(thresh=1)
frames = [dfDGEup, dfDGEdown,]
dfDGE = pd.concat(frames)
dfDGE
```
